```python
def cc_plot(input_file:str):
    G = nx.Graph()
    with open(input_file, 'r') as f:
        next(f)  # Skip the header row
        for line in f.readlines():
            edge = line.strip().split(';')
            G.add_edge(int(edge[0]), int(edge[1]))
    # Compute the clustering coefficient distribution
    cc = node_clustering_coefficient(G)
    # Plot the histogram of the distribution with interpolated curve
    counts, bin_edges = np.histogram(list(cc.values()), bins='auto')
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    # Interpolate the histogram tips with a curve
    x = np.linspace(min(bin_centers), max(bin_centers), 100)
    y = make_interp_spline(bin_centers, counts)(x)
    plt.plot(x, y, color='red')
    name = input_file.strip().split('.')[0].split('_')

    # Plot the histogram bars
    plt.hist(list(cc.values()), bins='auto', alpha=0.5)
    plt.xlabel('Clustering Coefficient')
    plt.ylabel('Count')
    plt.title(f'CC distribution of k = {name[2]}, c = {name[3]}')
    plt.show()
```

```python

def load_graph(input_file:str):
    G = nx.Graph()
    with open(input_file, 'r') as f:
        next(f)  # Skip the header row
        for line in f.readlines():
            edge = line.strip().split(';')
            G.add_edge(int(edge[0]), int(edge[1]))
    connectivity = nx.average_node_connectivity(G)
    print(f'Average node connectivity: {connectivity}')
    degree_sequence = [d for n, d in G.degree()]
    counts, bin_edges = np.histogram(degree_sequence, bins='auto')
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    name = input_file.strip().split('.')[0].split('_')
    # Interpolate the histogram tips with a curve
    x = np.linspace(min(bin_centers), max(bin_centers), 100)
    y = make_interp_spline(bin_centers, counts)(x)
    plt.plot(x, y, color='red')

    # Plot the histogram bars
    plt.hist(degree_sequence, bins='auto', alpha=0.5)
    plt.xlabel('Degree')
    plt.ylabel('Count')
    plt.title(f'Degree distribution of k = {name[2]}, c = {name[3]}')
    plt.show()
```
